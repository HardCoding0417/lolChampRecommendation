```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import StaleElementReferenceException



import pandas as pd
```

```python

# ChromeOptions 객체를 생성
options = webdriver.ChromeOptions()

# 미디어 스트림과 이미지 로딩 비활성화
options.add_argument("--disable-media-source")
options.add_argument("--blink-settings=imagesEnabled=false")
options.add_argument("--disable-javascript")

# WebDriver 설정
service = ChromeService(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 웹 페이지 로드
url = 'https://www.op.gg/champions/naafiri'
driver.get(url)
print('페이지 로드 완료')

# 광고와 관련된 요소 제거
driver.execute_script("var elems = document.querySelectorAll('.primisslate, .vm-placement'); for (var i = 0; i < elems.length; i++) { elems[i].remove(); }")
print('광고 요소 제거 완료')

# 버튼 클릭을 위한 XPath
button_xpath = "//button[@type='button' and @class='css-5e1aqy e1nhwouu1']"

# 현재 로딩된 마지막 댓글의 번호
current_last_comment = 0
while current_last_comment < 990:  # 최대 990번째 댓글까지 검사하면서 버튼 클릭
    try:
        print('버튼클릭')
        
        # 예외가 발생하면 최대 3번까지 다시 시도
        for _ in range(3):
            try:
                button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, button_xpath)))
                driver.execute_script("arguments[0].click();", button)
                print("클릭함")
                break
            except StaleElementReferenceException:
                continue

    except TimeoutException:
        print("더보기 버튼을 찾을 수 없습니다.")
        break

    # 새롭게 로딩된 마지막 댓글의 번호 검사
    for i in range(current_last_comment + 10, current_last_comment, -1):
        try:
            try:
                # 댓글 요소 확인
                comment_xpath = f'//*[@id="content-container"]/main/div[6]/div/div/div/div/div[2]/ul/li[{i}]/div[2]/p'
                driver.find_element(By.XPATH, comment_xpath)
            except:
                comment_xpath = f'//*[@id="content-container"]/main/div[7]/div/div/div/div/div[2]/ul/li[{i}]/div[2]/p'
                driver.find_element(By.XPATH, comment_xpath)
            current_last_comment = i
            break
        except NoSuchElementException:
            continue

# 1부터 1000까지의 댓글을 수집
texts = []
for i in range(1, 1001):
    try:
        try:
            element_xpath = f'//*[@id="content-container"]/main/div[6]/div/div/div/div/div[2]/ul/li[{i}]/div[2]/p'
            element = driver.find_element(By.XPATH, element_xpath)
        except:
            element_xpath = f'//*[@id="content-container"]/main/div[7]/div/div/div/div/div[2]/ul/li[{i}]/div[2]/p'
            element = driver.find_element(By.XPATH, element_xpath)
        texts.append(element.text)
    except NoSuchElementException:
        print(f"{i}번째 댓글이 존재하지 않습니다.")
        break

# 드라이버 종료
driver.quit()

# 텍스트 정보를 파일로 저장
file_path = "C:\\anaconda\\aiProject\\lol_project\\opgg_champ_url\\naafiri_repl.txt"
with open(file_path, 'w', encoding='utf-8') as file:
    for text in texts:
        file.write(text + '\n')
```
