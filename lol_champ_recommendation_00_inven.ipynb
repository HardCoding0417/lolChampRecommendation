```python
# ChromeOptions 객체를 생성
options = webdriver.ChromeOptions()

# 미디어 스트림과 이미지 로딩 비활성화
options.add_argument("--disable-media-source")
options.add_argument("--blink-settings=imagesEnabled=false")
options.add_argument("--disable-javascript")

# WebDriver 설정
service = ChromeService(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 웹 페이지 로드
driver.get('https://lol.inven.co.kr/dataninfo/champion/detail.php?code=19')

# 광고와 관련된 요소 제거
driver.execute_script("var elems = document.querySelectorAll('.primisslate, .vm-placement'); for (var i = 0; i < elems.length; i++) { elems[i].remove(); }")

xpath = '/html/body/div[4]/div[1]/section/article/section[2]/div[2]/div/div/div[3]/div[2]/div[2]/table/tbody/tr[1]/td[2]/span/span'
#        /html/body/div[4]/div[1]/section/article/section[2]/div[2]/div/div/div[3]/div[2]/div[2]/table/tbody/tr[2]/td[2]/span/span

element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, xpath))
)
element_txt = element.text
print(element_txt)
driver.quit()
```
